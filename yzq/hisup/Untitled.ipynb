```python
# import cv2
# import numpy as np
# import matplotlib.pyplot as plt

# # 
# image = cv2.imread('input_image.png', cv2.IMREAD_GRAYSCALE)

# 
if image is None:
    raise ValueError("")

#  [0, 1] 
image = image.astype(np.float32) / 255.0

# 
def gravitational_field(x, y, image, G=1):
    height, width = image.shape
    g_x = np.zeros_like(image)
    g_y = np.zeros_like(image)

    for i in range(height):
        for j in range(width):
            if i == x and j == y:
                continue  # 
            dx = x - i
            dy = y - j
            r = np.sqrt(dx**2 + dy**2)
            if r == 0:
                continue  # 
            g_x[x, y] += -G * image[i, j] * dx / r**3
            g_y[x, y] += -G * image[i, j] * dy / r**3

    return g_x, g_y

# 
height, width = image.shape
g_x = np.zeros((height, width))
g_y = np.zeros((height, width))

for i in range(height):
    for j in range(width):
        g_x[i, j], g_y[i, j] = gravitational_field(i, j, image)

# 
plt.figure(figsize=(10, 10))
plt.quiver(np.arange(width), np.arange(height), g_x, g_y, color='blue', scale=50, width=0.005)
plt.title('Gravitational Field (Vector Field)')
plt.xlabel('X')
plt.ylabel('Y')
plt.gca().invert_yaxis()  #  Y 
plt.grid(True)

# 
plt.savefig('gravitational_field.png', bbox_inches='tight')
plt.show()

# 
phi = -np.sqrt(g_x**2 + g_y**2)  # 
plt.figure(figsize=(10, 10))
contour = plt.contourf(np.arange(width), np.arange(height), phi, levels=50, cmap='viridis')
plt.colorbar(contour, label='Gravitational Potential')
plt.title('Gravitational Potential (Contour Plot)')
plt.xlabel('X')
plt.ylabel('Y')
plt.gca().invert_yaxis()  #  Y 
plt.grid(True)

# 
plt.savefig('gravitational_potential.png', bbox_inches='tight')
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

# 
if image is None:
    raise ValueError("")

#  [0, 1] 
image = image.astype(np.float32) / 255.0

# 
def gravitational_kernel(size):
    """"""
    x = np.arange(size) - size // 2
    y = np.arange(size) - size // 2
    xx, yy = np.meshgrid(x, y)
    r = np.sqrt(xx**2 + yy**2)
    r[r == 0] = 1  # 
    kernel = 1 / r**3
    return kernel, xx, yy

# 
kernel_size = 51  # 
kernel, xx, yy = gravitational_kernel(kernel_size)

# 
def compute_gravitational_field(image, kernel, xx, yy):
    """"""
    g_x = np.zeros_like(image)
    g_y = np.zeros_like(image)
    height, width = image.shape
    pad_size = kernel_size // 2

    # 
    image_padded = np.pad(image, pad_size, mode='constant')

    # 
    for i in range(height):
        for j in range(width):
            # 
            local_region = image_padded[i:i+kernel_size, j:j+kernel_size]
            # 
            g_x[i, j] = np.sum(local_region * kernel * xx)
            g_y[i, j] = np.sum(local_region * kernel * yy)

    return g_x, g_y

# 
g_x, g_y = compute_gravitational_field(image, kernel, xx, yy)

# 
plt.figure(figsize=(10, 10))
plt.quiver(np.arange(image.shape[1]), np.arange(image.shape[0]), g_x, g_y, color='blue', scale=50, width=0.005)
plt.title('Gravitational Field (Vector Field)')
plt.xlabel('X')
plt.ylabel('Y')
plt.gca().invert_yaxis()  #  Y 
plt.grid(True)
plt.savefig('gravitational_field.png', bbox_inches='tight')
plt.show()

# 
phi = -np.sqrt(g_x**2 + g_y**2)  # 
plt.figure(figsize=(10, 10))
contour = plt.contourf(np.arange(image.shape[1]), np.arange(image.shape[0]), phi, levels=50, cmap='viridis')
plt.colorbar(contour, label='Gravitational Potential')
plt.title('Gravitational Potential (Contour Plot)')
plt.xlabel('X')
plt.ylabel('Y')
plt.gca().invert_yaxis()  #  Y 
plt.grid(True)
plt.savefig('gravitational_potential.png', bbox_inches='tight')
plt.show()
```
